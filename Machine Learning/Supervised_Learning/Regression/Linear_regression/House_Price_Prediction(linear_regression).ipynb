<h1>House_Price_Prediction

<h3>Using Linear Regression

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
calif=fetch_california_housing(as_frame=True)
```

```python
calif
```

```python
df_x=pd.DataFrame(calif.data,columns=calif.feature_names)
```

```python
df_y=pd.DataFrame(calif.target)
```

```python
df_x.describe()
```

```python
reg=linear_model.LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
a=reg.predict(x_test)
```

```python
a
```

```python
y_test.head()
```

```python
np.mean(a-y_test**2)   #Mean squared Error
```

```python
reg.score(x_test,y_test)*100   #Accuracy
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,a)
```
