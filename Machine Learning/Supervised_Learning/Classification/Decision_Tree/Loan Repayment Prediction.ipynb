```python
#importing Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("C:\\Users\\Harsha\\Desktop\\Machine Learning\\Supervised_Learning\\Classification\\Decision_Tree\\Decision_Tree_ Dataset.csv")
```

```python
data.head()
```

```python
data=data.drop(columns="sum")
```

```python
df_x=data.values[:,:-1]
```

```python
df_x[0]
```

```python
df_x.shape
```

```python
df_y=data.values[:,-1]
```

```python
df_y[0]
```

```python
df_y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.3,random_state=50)
```

```python
#implementing the model
dec_classif=DecisionTreeClassifier(criterion="entropy",random_state=25,max_depth=10,min_samples_leaf=5)
```

```python
dec_classif.fit(x_train,y_train)
```

```python
y_pred=dec_classif.predict(x_test)
```

```python
#accuracy on test data
accuracy_score(y_test,y_pred)*100
```

```python
#testing accuracy on training data
y_train_pred=dec_classif.predict(x_train)
```

```python
accuracy_score(y_train,y_train_pred)*100
```
