```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets._samples_generator import make_blobs
```

```python
x,y=make_blobs(n_samples=40,centers=2,random_state=20)
```

```python
clf=svm.SVC(kernel="linear",C=1)
```

```python
clf.fit(x,y)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Accent_r)
plt.show()
```

```python
clf=svm.SVC(kernel="linear",C=1000)
clf.fit(x,y)
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Accent_r)
# Create a plot with limits based on the data range
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid to evaluate the model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY,XX = np.meshgrid(yy,xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

# Compute decision function
z = clf.decision_function(xy).reshape(XX.shape)

# Plot the decision boundary and margins
ax.contour(XX, YY, z, levels=[-1, 0, 1], colors='k', linestyles=['--', '-', '--'], alpha=0.5)

# Plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', linewidth=1,edgecolors="k")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Fit SVM model
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(x, y)

# Plot data points
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Accent_r)
# Create grid and evaluate model
ax = plt.gca()
xx, yy = np.meshgrid(np.linspace(*ax.get_xlim(), 30), np.linspace(*ax.get_ylim(), 30))
z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)


# Plot decision boundary and margins
ax.contour(xx, yy, z, levels=[-1, 0, 1], colors='k', linestyles=['--', '-', '--'], alpha=0.5)
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')

plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Fit SVM model
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(x, y)

# Plot decision regions
plot_decision_regions(X=x, y=y, clf=clf, legend=2)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')  # Plot support vectors
plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt

# Fit SVM model
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(x, y)

# Plot decision boundary
disp = DecisionBoundaryDisplay.from_estimator(clf, x, response_method="decision_function", alpha=0.5)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Accent_r)  # Plot data points
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')  # Plot support vectors
plt.show()
```
